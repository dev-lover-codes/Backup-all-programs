```python
import os
import subprocess
from google import genai
from google.genai import types

# --- 1. Define Real Tools ---
def read_file(file_path: str):
    """Reads a file from the user's project."""
    try:
        if not os.path.exists(file_path): return "File not found."
        with open(file_path, 'r') as f: return f.read()
    except Exception as e: return str(e)

def run_terminal(command: str):
    """Executes a terminal command (Sandboxed)."""
    # Safety: Block dangerous commands for this demo
    if "rm" in command or "sudo" in command: return "Command blocked for safety."

    try:
        result = subprocess.run(command, shell=True, capture_output=True, text=True)
        return result.stdout + result.stderr
    except Exception as e: return str(e)

# --- 2. Initialize Client with Tools ---
client = genai.Client(api_key="AIzaSyD5V9whqanBh_83TF8alaapCxSW-USLgvM")

# Create a chat session with tools enabled
chat = client.chats.create(
    model="gemini-2.0-flash",
    config=types.GenerateContentConfig(
        tools=[read_file, run_terminal], # Pass actual functions here!
        automatic_function_calling=types.AutomaticFunctionCallingConfig(
            disable=False # This tells Gemini to execute the Python functions for you
        ),
        system_instruction="You are an autonomous coding agent. You can read files and run commands to help the user."
    )
)

def run_agent(user_prompt):
    # The SDK handles the loop:
    # 1. Model decides to call tool -> 2. SDK runs function -> 3. SDK sends result back -> 4. Model replies
    response = chat.send_message(user_prompt)
    return response.text
```

```python
def run_voice_agent():
    print("Agent says: Hello, I am your text-based assistant. How can I help you?")
    while True:
        user_input_text = input("You say: ")
        if user_input_text.lower() == "exit":
            print("Agent says: Goodbye!")
            break
        if user_input_text:
            response_text = run_agent(user_input_text)
            print(f"Agent says: {response_text}")
```

```python
import speech_recognition as sr
from gtts import gTTS
from pydub import AudioSegment
from pydub.playback import play
import io

def recognize_speech():
    """Converts speech from the microphone to text."""
    r = sr.Recognizer()
    with sr.Microphone() as source:
        print("Say something!")
        r.adjust_for_ambient_noise(source)
        audio = r.listen(source)

    try:
        text = r.recognize_google(audio)
        print(f"You said: {text}")
        return text
    except sr.UnknownValueError:
        print("Google Speech Recognition could not understand audio")
        return ""
    except sr.RequestError as e:
        print(f"Could not request results from Google Speech Recognition service; {e}")
        return ""

def speak_text(text):
    """Converts text to speech and plays it."""
    print(f"Agent says: {text}")
    try:
        tts = gTTS(text=text, lang='en')
        fp = io.BytesIO()
        tts.write_to_fp(fp)
        fp.seek(0)
        audio = AudioSegment.from_file(fp, format="mp3")
        play(audio)
    except Exception as e:
        print(f"Error playing audio: {e}")

# Example usage (uncomment to test):
# speak_text("Hello, how can I help you today?")
# user_input = recognize_speech()
# if user_input:
#     speak_text(f"You just said: {user_input}")
```
